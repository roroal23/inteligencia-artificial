```python
import networkx as nx
import matplotlib.pyplot as plt
```

###Ejercicio 1

```python
G1 = nx.Graph()
G1.add_edge("1", "2", weight=200)
G1.add_edge("2", "3", weight=150)
G1.add_edge("2", "4", weight=350)
G1.add_edge("2", "5", weight=450)
G1.add_edge("3", "5", weight=400)
G1.add_edge("3", "6", weight=225)
G1.add_edge("4", "5", weight=300)
G1.add_edge("5", "7", weight=250)
G1.add_edge("7", "8", weight=125)
```

```python
G1.nodes
```

```python
G1.edges
```

```python
nx.draw(G1, with_labels=True)
```

```python
def heuristic_ejercicio1(current, target):
  h=0
  if (current=='1'): h = 800
  if (current=='2'): h = 650
  if (current=='3'): h = 500
  if (current=='4'): h = 650
  if (current=='5'): h = 325
  if (current=='6'): h = 275
  if (current=='7'): h = 125
  if (current == target): h = 0
  return h
```

```python
nx.astar_path(G1, "1", "8", heuristic = heuristic_ejercicio1, weight="weight")
```

```python
nx.astar_path_length (G1, "1", "8", heuristic = heuristic_ejercicio1, weight="weight")
```

###Ejercicio 2

```python
G2 = nx.Graph()
G2.add_edge("Coruña", "Oviedo", weight=325)
G2.add_edge("Coruña", "Salamanca", weight=410)
G2.add_edge("Coruña", "Madrid", weight=682)
G2.add_edge("Salamanca", "Oviedo", weight=312)
G2.add_edge("Salamanca", "Madrid", weight=230)
G2.add_edge("Oviedo", "Santander", weight=210)
G2.add_edge("Santander", "Burgos", weight=159)
G2.add_edge("Burgos", "Madrid", weight=239)
G2.add_edge("Burgos", "Zaragoza", weight=285)
G2.add_edge("Zaragoza", "Madrid", weight=322)
G2.add_edge("Zaragoza", "Barcelona", weight=299)
```

```python
def heuristic_ejercicio2(current, target):
  h=0
  if (current=='Coruña'): h = 1036
  if (current=='Burgos'): h = 577
  if (current=='Madrid'): h = 573
  if (current=='Oviedo'): h = 805
  if (current=='Salamanca'): h = 759
  if (current=='Santander'): h = 632
  if (current=='Zaragoza'): h = 282
  if (current == target): h = 0
  return h
```

```python
nx.astar_path(G2, "Coruña", "Barcelona", heuristic = heuristic_ejercicio2, weight="weight")
```

```python
nx.astar_path_length (G2, "Coruña", "Barcelona", heuristic = heuristic_ejercicio2, weight="weight")
```
